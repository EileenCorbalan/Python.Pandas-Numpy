14) ¿Cual es la expectativa de vida mas alta y mas baja? Usa la funcion de pandas y la funcion de NumPy y compara el resultado. ¿Que paises tienen esas expectativas de vida?

```python
# Maxima expectativa de vida usando numpy
# 14.1 Obtener la columna 'life_expectancy' como un arreglo de NumPy
life_expectancy = expvida['life_expectancy'].values

# 14.2 Obtener el valor máximo de expectativa de vida
maxexpvida_np = np.max(life_expectancy)
maxexpvida_np
```

```python
# 14.3 Obtener el valor máximo de expectativa de vida obviando los valores nulos
maxexpvida_nonan_np = np.nanmax(life_expectancy)
maxexpvida_nonan_np
```

```python
# 14.4 Obtener la maxima expectativa de vida usando pandas
maxexpvida_pd = expvida["life_expectancy"].max()
maxexpvida_pd
```

```python
# 14.5 Obtener la minima expectativa de vida usando numpy
minexpvida_np = np.min(life_expectancy)
minexpvida_np
```

```python
# 14.6 Obtener el valor mínimo de expectativa de vida obviando los valores nulos
minexpvida_nonan_np = np.nanmin(life_expectancy)
minexpvida_nonan_np
```

```python
# 14.7 Obtener la minima expectativa de vida usando pandas
minexpvida_pd = expvida["life_expectancy"].min()
minexpvida_pd
```

```python
# Mostrar los resultados maximos
print("Máxima expectativa de vida:", maxexpvida_pd)
```

```python
# Mostrar los resultados minimos
print("Mínima expectativa de vida:", minexpvida_pd )
```

```python
# Filtrar las columnas para mostrar solo las filas que tengan valor
# igual al maximo
filasmax = expvida[expvida["life_expectancy"] == maxexpvida_pd]
filasmax
```

```python
# Filtro las columnas para mostrar solo las filas que tengan valor
# igual al minimo
filasmin = expvida[expvida["life_expectancy"] == minexpvida_pd]
filasmin
```
