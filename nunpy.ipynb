Como dijimos en clases anteriores, Python tiene implementadas muchas librerias para poder trabajar con datos. En la clase de hoy trabajaremos con una de ellas: `Numpy`.

Antes de comenzar, vamos a hablar un poco de esta libreria o modulo.

**Numpy** es una librería optimizada para realizar cálculos numéricos con vectores y matrices. A diferencia de otros lenguajes de programación, Python no posee en su estructura central la figura de matrices. Eso quiere decir que para poder trabajar con esta estructura de datos deberiamos trabajar con listas de listas. NumPy introduce el concepto de arrays o matrices.

Al ser de código abierto, `numpy` posee una documentación muy amplia que es **SIEMPRE RECOMENDABLE** consultar.

```python
# Ejercicio 5
# Crear un arreglo de números aleatorios de longitud 100 y obtener su media y varianza
# 5.1 Crear el arreglo de números aleatorios
array5 = np.random.randint ( 0, 100, size=100)
array5
```

```python
# Ejercicio 6
# Calcular la media de un arreglo usando np.sum
# 6.1 Sumar el total de los elementos
sumarray3 = np.sum(array3)
sumarray3
```

```python
# 6.2 Obtener la cantidad de elementos
lenarray3 = len(array3)
lenarray3
```

```python
 # 6.3 Calcular la media
mediaarray3 = sumarray3 / lenarray3
mediaarray3
```

```python
#Ejercicio 7
# Calcular la varianza de un arreglo usando np.sum y np.mean
# 7.1 Calcular la media
meanarray3 = np.mean(array3)
meanarray3
```

```python
# 7.2 Obtener las desviaciones
desviaciones = array3 - meanarray3
desviaciones
```

```python
# 7.3 Elevar al cuadrado las desviaciones
desviacionescuadrado = desviaciones**2
desviacionescuadrado
```

```python
# 7.4 Calcular la varianza. Calcular la suma de las desviaciones al cuadrado y dividir por la cantidad de elementos
varianzaarray3 = np.sum(desviacionescuadrado) / len(array3)
varianzaarray3
```
